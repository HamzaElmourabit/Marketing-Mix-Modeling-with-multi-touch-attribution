```python
cols = df_raw.columns.tolist()
n = len(df_raw)
print("Analyzing columns without keyword assumptions...")

# 1) Detect datetime columns (dtype or parseable)
datetime_cols = [c for c in cols if np.issubdtype(df_raw[c].dtype, np.datetime64)]
for c in cols:
    if c in datetime_cols:
        continue
    if df_raw[c].dtype == object:
        parsed = pd.to_datetime(df_raw[c], errors='coerce')
        if parsed.notna().sum() / max(1, n) > 0.9:
            datetime_cols.append(c)

# 2) ID-like columns: very high cardinality
id_cols = [c for c in cols if df_raw[c].nunique(dropna=False) / max(1, n) > 0.9]

# 3) Numeric vs categorical
numeric_cols = df_raw.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = [c for c in cols if c not in numeric_cols + datetime_cols + id_cols]

# 4) Further numeric classification by simple statistics
stats = {}
for c in numeric_cols:
    s = df_raw[c]
    zero_frac = (s == 0).sum() / max(1, n)
    neg_frac = (s < 0).sum() / max(1, n)
    unique = s.nunique(dropna=True)
    is_binary = unique == 2 and set(s.dropna().unique()).issubset({0, 1})
    is_integer = np.allclose(s.dropna(), s.dropna().astype('int64')) if s.dropna().shape[0] > 0 else False
    stats[c] = dict(zero_frac=zero_frac, neg_frac=neg_frac, unique=unique, is_binary=is_binary, is_integer=is_integer,
                    mean=s.mean(), std=s.std())

# Heuristics (data-driven, not keyword-based):
# - Candidate media spends: numeric, non-negative, substantial fraction of zeros (sparse), not binary
# - Candidate KPI: numeric, not sparse zeros, positive mean
# - Controls: remaining numeric columns
candidate_media_spend = [c for c, v in stats.items() if v['neg_frac'] == 0 and v['zero_frac'] >= 0.1 and not v['is_binary']]
candidate_binary = [c for c, v in stats.items() if v['is_binary']]
candidate_kpi = [c for c, v in stats.items() if c not in candidate_media_spend + candidate_binary and v['mean'] > 0]
candidate_controls = [c for c in numeric_cols if c not in candidate_media_spend + candidate_kpi + candidate_binary]

print('\nSummary of detected column groups:')
print(f'  - datetime_cols ({len(datetime_cols)}): {datetime_cols}')
print(f'  - id_cols ({len(id_cols)}): {id_cols}')
print(f'  - categorical_cols ({len(categorical_cols)}): {categorical_cols[:20]}{"..." if len(categorical_cols)>20 else ""}')
print(f'  - numeric_cols ({len(numeric_cols)}): {numeric_cols[:20]}{"..." if len(numeric_cols)>20 else ""}')
print(f'  - candidate_media_spend ({len(candidate_media_spend)}): {candidate_media_spend}')
print(f'  - candidate_kpi ({len(candidate_kpi)}): {candidate_kpi}')
print(f'  - candidate_binary ({len(candidate_binary)}): {candidate_binary}')
print(f'  - candidate_controls ({len(candidate_controls)}): {candidate_controls}')

# Assign groups for downstream imputations
media_spend_cols = candidate_media_spend
kpi_cols = candidate_kpi
control_cols = candidate_controls

print('\nUse these groups for imputation or override manually if needed.')
```

```python
# ==============================================================================
# Phase 4 Corrigée : Gestion des valeurs manquantes (Sans bug d'alignement)
# ==============================================================================
import numpy as np
import pandas as pd

# Pour éviter la corruption de l'état, on repart toujours de df_raw initialement chargé
# Si df_raw a déjà été écrasé par des NaN, veuillez réexécuter la cellule de chargement du CSV.

media_cols = [c for c in media_spend_cols if c in df_raw.columns]
kpi_cols = [c for c in kpi_cols if c in df_raw.columns]
control_cols = [c for c in control_cols if c in df_raw.columns]
phase4_cols = media_cols + kpi_cols + control_cols

print('\nPhase 4 : Gestion des valeurs manquantes')
print(f'  - médias (fillna(0)) : {len(media_cols)} colonnes')
print(f'  - KPI : {len(kpi_cols)} colonnes')
print(f'  - contrôles continus : {len(control_cols)} colonnes')

before_report = df_raw[phase4_cols].isna().sum().rename('nan_before').reset_index()
before_report.columns = ['column', 'nan_before']
before_report['type'] = before_report['column'].apply(
    lambda c: 'média' if c in media_cols else ('KPI' if c in kpi_cols else 'contrôle continu')
)

if 'MMM_TIMESERIES_ID' in df_raw.columns:
    groups = df_raw.groupby('MMM_TIMESERIES_ID')
    out_frames = []
    for name, g in groups:
        g = g.sort_values(by=date_col).copy()

        # Imputation Médias (fillna(0))
        for col in media_cols:
            g[col] = g[col].fillna(0)

        # Imputation KPI & Contrôles (Interpolation linéaire préservant l'index)
        for col in (kpi_cols + control_cols):
            # Utilisation de .values ou interpolation directe pour éviter le désalignement d'index
            g[col] = g[col].interpolate(method='linear', limit_direction='both')
            g[col] = g[col].ffill().bfill()

        out_frames.append(g)

    df_clean = pd.concat(out_frames, axis=0).sort_values(by=[date_col, 'MMM_TIMESERIES_ID']).reset_index(drop=True)
else:
    df_clean = df_raw.sort_values(by=date_col).copy()
    for col in media_cols:
        df_clean[col] = df_clean[col].fillna(0)
    for col in (kpi_cols + control_cols):
        df_clean[col] = df_clean[col].interpolate(method='linear', limit_direction='both').ffill().bfill()

after_report = df_clean[phase4_cols].isna().sum().rename('nan_after').reset_index()
after_report.columns = ['column', 'nan_after']

report = before_report.merge(after_report, on='column')
report['filled'] = report['nan_before'] - report['nan_after']
report = report[['column', 'type', 'nan_before', 'nan_after', 'filled']].sort_values(
    by=['nan_before', 'nan_after', 'column'], ascending=[False, True, True]
)

print('\nRapport d\'imputation :')
print(report.to_string(index=False))

remaining = df_clean[phase4_cols].isna().sum()
remaining = remaining[remaining > 0]
if remaining.empty:
    print('\n✅ Succès : aucune valeur NaN restante dans les colonnes média, KPI et contrôles.')
else:
    print('\n⚠️ Échec : des NaN persistent :')
    print(remaining)

# Sauvegarde propre
df_raw = df_clean.copy()
```

```python
import logging
import time
import numpy as np
import pandas as pd

# Configuration du logging de production
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
logger = logging.getLogger(__name__)

try:
    start_time = time.time()

    # 1. Conversion de la colonne date en format datetime64[ns]
    date_col = "DATE_DAY"
    df_raw[date_col] = pd.to_datetime(df_raw[date_col], errors="coerce")
    
    # 2. Création de la colonne week_start (Lundi de la semaine ISO 8601)
    # Calcul vectorisé : soustraction du jour de la semaine (0=lundi, 6=dimanche)
    df_raw["week_start"] = df_raw[date_col] - pd.to_timedelta(df_raw[date_col].dt.weekday, unit="D")

    # 3. Classification dynamique des colonnes pour l'agrégation
    group_keys = ["MMM_TIMESERIES_ID", "week_start"]
    
    # Métadonnées catégorielles à conserver
    metadata_candidates = [
        "ORGANISATION_ID",
        "ORGANISATION_VERTICAL",
        "ORGANISATION_SUBVERTICAL",
        "ORGANISATION_MARKETING_SOURCES",
        "ORGANISATION_PRIMARY_TERRITORY_NAME",
        "TERRITORY_NAME",
        "CURRENCY_CODE"
    ]
    dims_to_keep = [col for col in metadata_candidates if col in df_raw.columns]

    # Colonnes à sommer (Somme)
    spend_cols = [col for col in df_raw.columns if "_SPEND" in col]
    impression_cols = [col for col in df_raw.columns if "_IMPRESSIONS" in col]
    click_cols = [col for col in df_raw.columns if "_CLICKS" in col]
    purchase_cols = [
        "FIRST_PURCHASES", "FIRST_PURCHASES_UNITS",
        "ALL_PURCHASES", "ALL_PURCHASES_UNITS"
    ]
    sum_cols = list(set(spend_cols + impression_cols + click_cols + purchase_cols))
    sum_cols = [col for col in sum_cols if col in df_raw.columns]

    # Colonnes à moyenner (Moyenne)
    price_cols = [col for col in df_raw.columns if "_PRICE" in col or "_DISCOUNT" in col]
    weather_keywords = ["temp", "weather", "rain", "precip", "wind", "meteo", "humidity"]
    weather_cols = [
        col for col in df_raw.columns 
        if any(kw in col.lower() for kw in weather_keywords)
    ]
    mean_cols = list(set(price_cols + weather_cols))
    mean_cols = [col for col in mean_cols if col in df_raw.columns]

    # Construction du dictionnaire d'agrégation
    agg_dict = {}
    for col in sum_cols:
        agg_dict[col] = "sum"
    for col in mean_cols:
        agg_dict[col] = "mean"
    for col in dims_to_keep:
        agg_dict[col] = "first"

    # 4. Agrégation hebdomadaire
    df_weekly = df_raw.groupby(group_keys, as_index=False).agg(agg_dict)

    # 5. Vérification d'unicité des semaines par série temporelle
    duplicates = df_weekly.duplicated(subset=["MMM_TIMESERIES_ID", "week_start"]).sum()
    if duplicates > 0:
        raise ValueError(
            f"Doublons de granularité détectés : {duplicates} clés (MMM_TIMESERIES_ID, week_start) dupliquées."
        )
    logger.info("✅ Vérification d'unicité réussie : aucune semaine dupliquée.")

    # 6. Tri chronologique et réinitialisation de l'index
    df_weekly = df_weekly.sort_values(by=["MMM_TIMESERIES_ID", "week_start"]).reset_index(drop=True)

    # 7. Optimisation de la mémoire : Downcast des colonnes float64 en float32
    float64_cols = df_weekly.select_dtypes(include=["float64"]).columns
    df_weekly[float64_cols] = df_weekly[float64_cols].astype("float32")

    duration = time.time() - start_time
    logger.info(
        f"Pipeline terminé avec succès. Durée d'exécution : {duration:.4f}s. "
        f"Lignes traitées : {len(df_weekly)}."
    )

except Exception as e:
    logger.error(f"Échec lors de l'exécution de la Phase 2 : {str(e)}")
    raise
```

```python
# ==============================================================================
# Étape 3 : Isolation des Valeurs Aberrantes (Outliers) et Événements Exceptionnels
# ==============================================================================
import pandas as pd
import numpy as np
import logging

# Configuration du logging
logger = logging.getLogger(__name__)

try:
    logger.info("Début de la création des variables muettes événementielles...")
    
    # 1. Copie de sauvegarde
    df_events = df_weekly.copy()
    df_events["week_start"] = pd.to_datetime(df_events["week_start"])
    
    # 2. Définition des Confinements (Lockdowns) en France
    # Lockdown 1 : 17 mars 2020 au 11 mai 2020 (semaines du 16-03 au 11-05 inclus -> 9 semaines)
    df_events["is_lockdown_1"] = (
        (df_events["week_start"] >= "2020-03-16") & 
        (df_events["week_start"] <= "2020-05-11")
    ).astype(np.int8)
    
    # Lockdown 2 : 30 octobre 2020 au 15 décembre 2020 (semaines du 26-10 au 14-12 inclus -> 8 semaines)
    df_events["is_lockdown_2"] = (
        (df_events["week_start"] >= "2020-10-26") & 
        (df_events["week_start"] <= "2020-12-14")
    ).astype(np.int8)
    
    # Lockdown 3 : 3 avril 2021 au 3 mai 2021 (semaines du 29-03 au 26-04 inclus -> 5 semaines)
    df_events["is_lockdown_3"] = (
        (df_events["week_start"] >= "2021-03-29") & 
        (df_events["week_start"] <= "2021-04-26")
    ).astype(np.int8)
    
    # 3. Définition des événements annuels (Black Friday et Noël sur 2019-2023)
    bf_weeks = pd.to_datetime([
        "2019-11-25", "2020-11-23", "2021-11-22", "2022-11-21", "2023-11-20"
    ])
    df_events["is_black_friday"] = df_events["week_start"].isin(bf_weeks).astype(np.int8)
    
    christmas_weeks = pd.to_datetime([
        "2019-12-23", "2020-12-21", "2021-12-20", "2022-12-19", "2023-12-25"
    ])
    df_events["is_christmas"] = df_events["week_start"].isin(christmas_weeks).astype(np.int8)
    
    # 4. Définition des Soldes (Hiver et Été)
    # Soldes d'hiver : 6 semaines à partir du 2ème mercredi de janvier (ou 1er si après le 12)
    soldes_hiver_weeks = []
    for year in range(2020, 2025):
        jan1 = pd.Timestamp(year, 1, 1)
        first_wed = jan1 + pd.Timedelta(days=(2 - jan1.weekday()) % 7)
        second_wed = first_wed + pd.Timedelta(weeks=1)
        start_date = first_wed if second_wed.day > 12 else second_wed
        for w in range(6):
            soldes_hiver_weeks.append(start_date + pd.Timedelta(weeks=w))
            
    soldes_hiver_mondays = pd.to_datetime([d - pd.Timedelta(days=d.weekday()) for d in soldes_hiver_weeks])
    df_events["is_soldes_hiver"] = df_events["week_start"].isin(soldes_hiver_mondays).astype(np.int8)
    
    # Soldes d'été : 6 semaines à partir du dernier mercredi de juin
    # En 2019, on prend 7 semaines pour inclure le lundi 5 août (dernière semaine contenant le mardi 6 août fin de soldes)
    soldes_ete_weeks = []
    for year in range(2019, 2024):
        june30 = pd.Timestamp(year, 6, 30)
        start_date = june30 - pd.Timedelta(days=(june30.weekday() - 2) % 7)
        weeks_count = 7 if year == 2019 else 6
        for w in range(weeks_count):
            soldes_ete_weeks.append(start_date + pd.Timedelta(weeks=w))
            
    soldes_ete_mondays = pd.to_datetime([d - pd.Timedelta(days=d.weekday()) for d in soldes_ete_weeks])
    soldes_ete_mondays = soldes_ete_mondays[soldes_ete_mondays >= "2019-07-15"] # Filtrage du début de dataset
    df_events["is_soldes_ete"] = df_events["week_start"].isin(soldes_ete_mondays).astype(np.int8)
    
    # 5. Définition de la Rentrée scolaire (2 semaines par an, fin août / début septembre)
    rentree_weeks = pd.to_datetime([
        "2019-08-26", "2019-09-02",
        "2020-08-24", "2020-08-31",
        "2021-08-23", "2021-08-30",
        "2022-08-22", "2022-08-29",
        "2023-08-28", "2023-09-04"
    ])
    df_events["is_rentree"] = df_events["week_start"].isin(rentree_weeks).astype(np.int8)
    
    # 6. Validation statistique (sur les semaines uniques pour éviter la duplication par série ID)
    df_unique_weeks = df_events.drop_duplicates(subset=["week_start"])
    total_weeks = len(df_unique_weeks)
    
    event_cols = [
        "is_lockdown_1", "is_lockdown_2", "is_lockdown_3",
        "is_black_friday", "is_christmas", "is_soldes_hiver",
        "is_soldes_ete", "is_rentree"
    ]
    
    validation_df = pd.DataFrame(index=event_cols)
    validation_df["nb_semaines_flaggées"] = [df_unique_weeks[col].sum() for col in event_cols]
    validation_df["% du total"] = (validation_df["nb_semaines_flaggées"] / total_weeks * 100).round(1)
    
    print(f"📊 Semaines uniques dans le dataset : {total_weeks}")
    print("\n--- Validation des variables muettes ---")
    print(validation_df)
    
    # Mettre à jour df_weekly
    df_weekly = df_events
    logger.info("✅ Variables muettes créées et intégrées avec succès à df_weekly !")

except Exception as e:
    logger.error(f"❌ Échec lors de la création des variables muettes : {str(e)}")
    raise
```

```python
# ============================================================
# ÉTAPE 4.1 — Diagnostic VIF sur les colonnes spend granulaires
# ============================================================
import logging
import time
import numpy as np
import pandas as pd
from numpy.linalg import lstsq

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
logger = logging.getLogger(__name__)


def compute_vif_numpy(df: pd.DataFrame, feature_cols: list) -> pd.DataFrame:
    """VIF_i = 1 / (1 - R_i²) — implémentation pure NumPy."""
    cols_available = [c for c in feature_cols if c in df.columns]
    X = df[cols_available].fillna(0).values.astype(np.float64)
    n, k = X.shape

    records = []
    for i in range(k):
        y_i = X[:, i]
        X_others = np.delete(X, i, axis=1)
        X_with_const = np.column_stack([np.ones(n), X_others])
        coefs, _, _, _ = lstsq(X_with_const, y_i, rcond=None)
        y_pred = X_with_const @ coefs
        ss_res = np.sum((y_i - y_pred) ** 2)
        ss_tot = np.sum((y_i - y_i.mean()) ** 2)
        r2 = float(np.clip(1.0 - (ss_res / ss_tot) if ss_tot > 1e-10 else 0.0, 0.0, 1.0 - 1e-10))
        vif = 1.0 / (1.0 - r2)
        severity = "🔴 CRITIQUE (VIF ≥ 10)" if vif >= 10 else ("🟠 ÉLEVÉ (VIF ≥ 5)" if vif >= 5 else "🟢 OK (VIF < 5)")
        records.append({"feature": cols_available[i], "vif": round(vif, 2), "r2": round(r2, 4), "severity": severity})

    return pd.DataFrame(records).sort_values("vif", ascending=False).reset_index(drop=True)


spend_granular_cols = [
    "GOOGLE_PAID_SEARCH_SPEND", "GOOGLE_SHOPPING_SPEND", "GOOGLE_PMAX_SPEND",
    "GOOGLE_DISPLAY_SPEND", "GOOGLE_VIDEO_SPEND",
    "META_FACEBOOK_SPEND", "META_INSTAGRAM_SPEND", "META_OTHER_SPEND", "TIKTOK_SPEND",
]

vif_before = compute_vif_numpy(df_weekly, spend_granular_cols)

print("\n" + "=" * 65)
print("  DIAGNOSTIC VIF — Colonnes Spend Granulaires (avant agrégation)")
print("=" * 65)
print(vif_before.to_string(index=False))
print("=" * 65)
print(f"  Canaux critiques (VIF > 10) : {(vif_before['vif'] > 10).sum()}")
print(f"  Canaux élevés   (VIF > 5)  : {(vif_before['vif'] > 5).sum()}")
```

```python
# ============================================================
# ÉTAPE 4.2 — Agrégation des canaux corrélés en variables macro
# ============================================================

CHANNEL_GROUPS = {
    "PAID_SEARCH_SPEND":   ["GOOGLE_PAID_SEARCH_SPEND", "GOOGLE_SHOPPING_SPEND", "GOOGLE_PMAX_SPEND"],
    "DISPLAY_VIDEO_SPEND": ["GOOGLE_DISPLAY_SPEND", "GOOGLE_VIDEO_SPEND"],
    "PAID_SOCIAL_SPEND":   ["META_FACEBOOK_SPEND", "META_INSTAGRAM_SPEND", "META_OTHER_SPEND", "TIKTOK_SPEND"],
}

try:
    df_silver = df_weekly.copy()

    for macro_col, source_cols in CHANNEL_GROUPS.items():
        cols_present = [c for c in source_cols if c in df_silver.columns]
        cols_absent  = [c for c in source_cols if c not in df_silver.columns]

        if not cols_present:
            logger.warning(f"Aucune colonne source pour '{macro_col}' — ignoré.")
            continue

        # Somme vectorisée — NaN = 0 (canal absent = pas de dépense)
        df_silver[macro_col] = df_silver[cols_present].fillna(0).sum(axis=1)
        df_silver.drop(columns=cols_present, inplace=True)

        logger.info(f"✅ '{macro_col}' créée — sources : {cols_present}")
        if cols_absent:
            logger.warning(f"   ⚠️  Colonnes absentes ignorées : {cols_absent}")

    print(f"\n✅ Agrégation terminée — Shape final : {df_silver.shape}")
    spend_macro_cols = [c for c in CHANNEL_GROUPS.keys() if c in df_silver.columns]
    print(f"   Variables macro créées : {spend_macro_cols}")
    print(f"   Colonnes restantes     : {df_silver.columns.tolist()}")

except Exception as e:
    logger.error(f"Échec agrégation : {str(e)}")
    raise
```

```python
# ==============================================================================
# >>> AGENT : ÉTAPE 1 (RATIOS) & ÉTAPE 2 (ADSTOCK & SATURATION) <<<
# ==============================================================================
import logging
import numpy as np
import pandas as pd

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
logger = logging.getLogger(__name__)

try:
    logger.info("--- ÉTAPE 1 : CRÉATION DES RATIOS ET DÉCORRÉLATION (Idempotent) ---")
    
    # 1. Taux de remise (Discount Rate) sans division par zéro - Idempotent
    if 'ALL_PURCHASES_DISCOUNT_RATE' not in df_silver.columns:
        if 'ALL_PURCHASES_ORIGINAL_PRICE' in df_silver.columns and 'ALL_PURCHASES_GROSS_DISCOUNT' in df_silver.columns:
            df_silver['ALL_PURCHASES_DISCOUNT_RATE'] = np.where(
                df_silver['ALL_PURCHASES_ORIGINAL_PRICE'] > 0,
                df_silver['ALL_PURCHASES_GROSS_DISCOUNT'] / df_silver['ALL_PURCHASES_ORIGINAL_PRICE'],
                0.0
            )
            logger.info("✅ ALL_PURCHASES_DISCOUNT_RATE calculée.")
        else:
            logger.warning("⚠️ Colonnes brutes de prix/remise manquantes. Impossible de calculer le Discount Rate.")
            df_silver['ALL_PURCHASES_DISCOUNT_RATE'] = 0.0
    else:
        logger.info("ℹ️ ALL_PURCHASES_DISCOUNT_RATE déjà présente dans df_silver.")
    
    # 2. Prix moyen de vente (Average Selling Price) sans division par zéro - Idempotent
    if 'ALL_PURCHASES_AVERAGE_PRICE' not in df_silver.columns:
        if 'ALL_PURCHASES_UNITS' in df_silver.columns:
            # Si les colonnes brutes ont été supprimées mais qu'on a déjà tourné une fois, on ne recalcule pas
            if 'ALL_PURCHASES_ORIGINAL_PRICE' in df_silver.columns:
                df_silver['ALL_PURCHASES_AVERAGE_PRICE'] = np.where(
                    df_silver['ALL_PURCHASES_UNITS'] > 0,
                    df_silver['ALL_PURCHASES_ORIGINAL_PRICE'] / df_silver['ALL_PURCHASES_UNITS'],
                    0.0
                )
                logger.info("✅ ALL_PURCHASES_AVERAGE_PRICE calculée.")
            else:
                logger.warning("⚠️ ALL_PURCHASES_ORIGINAL_PRICE manquante. Impossible de calculer l'Average Price.")
                df_silver['ALL_PURCHASES_AVERAGE_PRICE'] = 0.0
        else:
            logger.warning("⚠️ ALL_PURCHASES_UNITS manquante. Impossible de calculer l'Average Price.")
            df_silver['ALL_PURCHASES_AVERAGE_PRICE'] = 0.0
    else:
        logger.info("ℹ️ ALL_PURCHASES_AVERAGE_PRICE déjà présente dans df_silver.")
    
    # 3. Suppression des colonnes brutes hautement colinéaires
    cols_to_drop = ['ALL_PURCHASES_ORIGINAL_PRICE', 'ALL_PURCHASES_GROSS_DISCOUNT']
    cols_present_to_drop = [c for c in cols_to_drop if c in df_silver.columns]
    if cols_present_to_drop:
        df_silver.drop(columns=cols_present_to_drop, inplace=True)
        logger.info(f"✅ Suppression des colonnes brutes : {cols_present_to_drop}")
    else:
        logger.info("ℹ️ Colonnes brutes déjà supprimées ou absentes.")

except Exception as e:
    logger.error(f"❌ Échec Étape 1 : {str(e)}")
    raise

try:
    logger.info("--- ÉTAPE 2 : TRANSFORMATION ADSTOCK ET SATURATION DE HILL (Idempotent) ---")
    
    # Import des modules d'adstock et de saturation du projet
    import sys
    import os
    project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
    if project_root not in sys.path:
        sys.path.insert(0, project_root)
        
    from transforms.adstock.GeometricAdstock import GeometricAdstock
    from transforms.saturation.HillSaturation import HillSaturation
    
    # Assurer le tri temporel par série pour l'adstock
    df_silver.sort_values(by=['MMM_TIMESERIES_ID', 'week_start'], inplace=True)
    
    spend_macro_cols = ["PAID_SEARCH_SPEND", "DISPLAY_VIDEO_SPEND", "PAID_SOCIAL_SPEND"]
    
    # Paramètres d'adstock (decay hebdomadaire)
    adstock_decays = {
        "PAID_SEARCH_SPEND": 0.3,
        "DISPLAY_VIDEO_SPEND": 0.4,
        "PAID_SOCIAL_SPEND": 0.35
    }
    
    for col in spend_macro_cols:
        if col in df_silver.columns:
            # La meilleure pratique pour l'idempotence de la cellule est de re-charger le spend depuis df_weekly si présent
            if 'df_weekly' in globals() and col in df_weekly.columns:
                raw_spend = df_weekly[col]
                logger.info(f"ℹ  Réinitialisation du spend brut pour {col} à partir de df_weekly.")
            else:
                raw_spend = df_silver[col]
                logger.warning(f"⚠️ Utilisation du spend de df_silver pour {col} (risque de double application si exécuté plusieurs fois sans relancer le notebook).")
            
            # 1. Adstock par série temporelle
            decay = adstock_decays.get(col, 0.3)
            adstock_model = GeometricAdstock(decay=decay)
            
            # Travailler sur les valeurs brutes réinitialisées pour l'idempotence
            df_temp = df_silver.copy()
            df_temp[col] = raw_spend
            
            adstocked = df_temp.groupby('MMM_TIMESERIES_ID')[col].transform(
                lambda x: adstock_model.transform(x.tolist())
            )
            
            # 2. Saturation de Hill (K dynamique = moyenne des spends non nuls réels)
            non_zero = df_temp[df_temp[col] > 0][col]
            K = float(non_zero.mean()) if not non_zero.empty else 100.0
            n = 1.5
            
            hill_model = HillSaturation(K=K, n=n)
            df_silver[col] = pd.Series(hill_model.transform(adstocked.tolist()), index=df_silver.index)
            
            logger.info(f"✅ {col} transformée : Adstock (decay={decay}) -> Saturation Hill (K={K:.2f}, n={n})")
            
except Exception as e:
    logger.error(f"❌ Échec Étape 2 : {str(e)}")
    raise
```

```python

import json
import os
import pandas as pd
import numpy as np

# 1. Définition des colonnes à scaler dans df_silver
spend_macro_cols = ["PAID_SEARCH_SPEND", "DISPLAY_VIDEO_SPEND", "PAID_SOCIAL_SPEND"]
kpi_cols = ["FIRST_PURCHASES", "ALL_PURCHASES", "FIRST_PURCHASES_UNITS", "ALL_PURCHASES_UNITS"]

# Les variables muettes (dummy) comme is_lockdown, is_christmas ne doivent PAS être scalées.
cols_to_scale = spend_macro_cols + kpi_cols
df_scaled = df_silver.copy()

scaling_metadata = {
    "scaling_method": {
        "spends": "mean_scaling",  # Robustesse aux valeurs aberrantes
        "kpis": "mean_scaling"     # Préservation de l'échelle linéaire et du zéro absolu
    },
    "parameters": {}
}

# 2. Calcul et application du scaling
for col in cols_to_scale:
    if col in df_scaled.columns:
        col_mean = float(df_scaled[col].mean())
        col_max = float(df_scaled[col].max())
        col_min = float(df_scaled[col].min())
        
        # Sauvegarde des paramètres pour l'inférence ou l'inversion
        scaling_metadata["parameters"][col] = {
            "mean": col_mean,
            "max": col_max,
            "min": col_min
        }
        
        # Application du scaling (Mean Scaling)
        # Formule : x_scaled = x / mean
        if col_mean > 0:
            df_scaled[col] = df_scaled[col] / col_mean
            print(f"✅ {col} scale par sa moyenne (mean = {col_mean:.2f})")
        else:
            print(f"⚠️ {col} possède une moyenne de 0 ou négative, scaling ignoré.")

# 3. Exportation des métadonnées de scaling en JSON
metadata_dir = "../data/metadata"
os.makedirs(metadata_dir, exist_ok=True)
metadata_path = os.path.join(metadata_dir, "scaling_metadata.json")

with open(metadata_path, "w", encoding="utf-8") as f:
    json.dump(scaling_metadata, f, ensure_ascii=False, indent=4)

print(f"\n💾 Métadonnées de scaling sauvegardées dans : {metadata_path}")

# Aperçu du dataset mis à l'échelle
print("\nStatistiques descriptives post-scaling (les moyennes doivent valoir 1.0) :")
print(df_scaled[cols_to_scale].describe().loc[['mean', 'min', 'max']])

# 4. Fonction utilitaire pour repasser à l'échelle réelle (inversion)
def inverse_transform(scaled_values, col_name, metadata_file=metadata_path):
    with open(metadata_file, "r") as f:
        meta = json.load(f)
    mean_val = meta["parameters"][col_name]["mean"]
    return scaled_values * mean_val

# Mettre à jour df_silver
df_silver_scaled = df_scaled
```
